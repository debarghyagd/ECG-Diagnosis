```python
import requests
import json
with open("./raw_signal.txt",'r') as f:
    L = [i.lstrip("'").rstrip("'") for i in f.readline().strip('[').strip(']\n').split(', ')]
#     print(L)
t=0
X=""
while(True):
    for i in range(len(L)):
        X={'time': t,'ECG': float(L[i])}#",{'time': "+str(t+0.001)+", 'ECG': "+str(L[i])+"},"
    #     if i%5==4:
    #     X="{"+X[:-1]+"}"
        X=json.dumps(X)
        print(X)
        resp = requests.post("http://192.168.189.158:3000/ecg-data",data=X, headers={'Content-type': 'application/json'})
        print(resp.status_code)
    #     X={'time': t+0.001,'ECG': L[i]}#",{'time': "+str(t+0.001)+", 'ECG': "+str(L[i])+"},"
    #     if i%5==4:
    #     X="{"+X[:-1]+"}"
    #     X=json.dumps(X)
    #     print(X)
    #     resp = requests.post("http://192.168.0.100:3000/pulse-data",data=X, headers={'Content-type': 'application/json'})
    #     print(resp.status_code)
        t+=0.002
        X=""
        print()
```
